```python
import matplotlib.pyplot
%matplotlib inline
```

```python
image  = matplotlib.pyplot.imshow(data)
matplotlib.pyplot.axes().set_aspect('auto')
matplotlib.pyplot.savefig('fig/01-numpy-heatmap-large.png')
matplotlib.pyplot.show(image)
```

```python
image  = matplotlib.pyplot.imshow(data)
matplotlib.pyplot.savefig('fig/01-numpy-heatmap-small.png')
matplotlib.pyplot.show(image)
```

```python
ave_temp = data.mean(axis=0)
max_temp = data.max(axis=0)
ave_plot = matplotlib.pyplot.plot(ave_temp,max_temp)
# matplotlib.pyplot.savefig('fig/01-numpy-avg_temp.png')
matplotlib.pyplot.show(ave_plot)
```

```python
fig = matplotlib.pyplot.figure(figsize=(5.0, 9.0))

axes1 = fig.add_subplot(4, 1, 1)
axes2 = fig.add_subplot(4, 1, 2)
axes3 = fig.add_subplot(4, 1, 3)
axes4 = fig.add_subplot(4, 1, 4)

axes1.set_ylabel('station 0')
axes1.plot(data[0,:])

axes2.set_ylabel('station 1')
axes2.plot(data[1,:])

axes3.set_ylabel('station 2')
axes3.plot(data[2,:])

axes4.set_ylabel('station 3')
axes4.plot(data[3,:])

matplotlib.pyplot.savefig('fig/01-numpy-subplots.png')
matplotlib.pyplot.show(fig)
```
